# Полиморфизм

```python
class Auto:
    def __init__(self, brand, model, power):
        self.brand = brand
        self.model = model
        self._power = power

    def get_tax(self):
        rate = 12
        if self._power > 100:
            rate = 25
        return self._power * rate

    def __str__(self):
        return f"{self.brand} {self.model}"

    def __repr__(self):
        return f"{self.brand} {self.model}"


class Bus(Auto):
    """
    Класс автобуса. Унаследован от Auto.
    """

    def set_places(self, seating, standing=0):
        """
        Метод для установки мест в автобусе:
            seating - седячие
            standing - стоячие
        """
        self.seating = seating
        self.standing = standing

    def get_total_places(self):
        """
        Метод для получения общего количества мест в автобусе.
        """
        return self.seating + self.standing

    def get_tax(self):
        """
        Переопределеяем метод get_tax()
        """
        rate = 15
        return rate * self._power


class Dealer:
    def __init__(self):
        self.cars = []

    def add_car(self, car):
        self.cars.append(car)

    def calc_total_tax(self):
        """
        Расчет налога для всего парка автомобилей.
        """
        tax = 0
        for car in self.cars:
            tax += car.get_tax()
        return tax


# Создаем автомобили
bmw = Auto("BMW", "X5", 400)
jt2 = Auto("Jetour", "T2", 256)

# Создаем автобусы
pazik = Bus("ПАЗ", "3204", 400)

# Отправляем автомобили и автобусы дилеру
dealer = Dealer()
dealer.add_car(bmw)
dealer.add_car(jt2)
dealer.add_car(pazik)

# Вводим автомобили в наличии
print(dealer.cars)

# Выводим суммарный налог
print(dealer.calc_total_tax())
```
